```python
def two_norm(x, h1, m1, s1, h2, m2, s2):
    return (
        h1*scipy.stats.norm.pdf(x, loc=m1 ,scale=s1)
        + h2*scipy.stats.norm.pdf(x, loc=m2 ,scale=s2)
    )

def density_est(x, x_vec, w, n=4):
    i = np.searchsorted(x_vec, x)
    w = x_vec[i+n] - x_vec[i-n+1]
    d = np.logical_and(np.greater_equal(x_vec, x-w/2), np.less_equal(x_vec, x+w/2))
    return sum(d)/w
```

```python
x_vec = np.linspace(10,100,1000)
x_sorted = np.array(sorted(data.data[MiNg_$150_PJM-W_2035]))
d_vec = [density_est(x, x_sorted, 1, n=250) for x in x_vec]
p, covar = scipy.optimize.curve_fit(two_norm, x_vec, d_vec, p0=[1000, 40, 3, 1000, 50, 10])

y_vec = two_norm(x_vec, *p)
```
